```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Migration_Success', axis=1)
y = df['Migration_Success']   
```

```python
    X.shape
```

```python
X_TRAIN, X_TEST, Y_TRAIN, Y_TEST = train_test_split(X, y, test_size=0.25, random_state=30)
```

```python
print("Size of Train X", len(X_TRAIN))
print("Size of Train y", len(Y_TRAIN))
print("Size of Test y", len(X_TEST))
print("Size of Test Y", len(Y_TEST))
```

```python
# Linear regression
# Machine learning
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(X_TRAIN, Y_TRAIN)
```

```python
#accuracy for prediction on traning dat

training_data_prediction = model.predict(X_TRAIN)
print(training_data_prediction)
```

```python
from sklearn.metrics import(
mean_absolute_error, mean_squared_error, r2_score, 
mean_absolute_percentage_error
)

r2 = r2_score(Y_TRAIN, training_data_prediction)
mea = mean_absolute_error(Y_TRAIN, training_data_prediction)
mse = mean_squared_error (Y_TRAIN, training_data_prediction)
rmse = np.sqrt(mse)
```

```python
print(f"R2 : {r2 : .4f}")
print (f"MAE : {mea: .4f}")
print(f"MSE : {mse : .4f}")
print(f"RMSE : {rmse : .4f}")
```

```python
y_pred = model.predict(X_TEST)
```

```python
r2 = r2_score(Y_TEST, y_pred)
mea = mean_absolute_error(Y_TEST, y_pred)
mse = mean_squared_error (Y_TEST, y_pred)
rmse = np.sqrt(mse)
```
